**Steady-state response:** Response of the system when time approaches infinity.

**Why steady-state response is important?**

It provides an indication of **accuracy** of a control system when its **output** is compared with the **desired input**. 

The steady-state response can be obtained by using **final value theorem.**





**Final Value Theorem (FVT)**

Finding the $y(\infty)$ is impossible mathematically. So, we can find the final value by computing the limits when the time approaches infinity.

$$\lim_{x\to\infty} y(t) = \lim_{s\to0} sY(s)$$

*   If the output converges to a single value, the final value exits.
*   If the output diverges to infinity or oscillates continuously, no meaning for the final value.

**Steady-State Error**

If the output and desired input do not agree exactly, then a **steady-state error** exists.

<figure>
<img src='https://live.staticflickr.com/65535/49982682857_c5d1bffb07_b.jpg' />
<figcaption> $\hspace{7cm}$ Figure 4</figcaption>

Any physical control system inherently suffers steady-state error in response to certain types of inputs.

The steady-state response of a system is judged by the **steady-state error** due to **step**, **ramp**, and **parabolic inputs**.

A system may have no steady-state error to a step input, but the sama system may exhibit nonzero steady-state error to a ramp input.


These inputs may be associated with the ability of a control system to:

*   Position itself relative to a stationary target
*   Follow a target moving at constant speed
*   Track an object that is accelerating


Consider the **unity-feedback** control system with the following **open-loop transfer function** :



<figure>
<img src='https://live.staticflickr.com/65535/49983842273_e2ce2910e4_b.jpg' />
<figcaption> $\hspace{7cm}$ Figure 5</figcaption>




$\hspace{2cm}G(s)=\frac{K(T_as+1)(T_bs+1)...(T_ms+1)}{s^N(T_1s+1)(T_2s+1)...(T_ps+1)}$

where $s^N$ in the denominator, represents N poles at the origin.



*   A system is called **type 0**, **type 1**, **type2**, ...,

  if $N=0$, $N=1$, $N=2$ ... 

For example:

$G(s)=\frac{s+1}{s}\hspace{0.5cm}$ is a type 1 system

$G(s)=\frac{s-3}{s^2}\hspace{0.5cm}$ is a type 2 system

$G(s)=\frac{s+1}{s+2}\hspace{0.5cm}$ is a type 0 system

The transfer function between the error signal $E(s)$ and the input signal $R(s)$ is:


$$\frac{E(s)}{R(s)}=\frac{1}{1+G(s)}$$

The final-value theorem provides a convenient way to find the **steady-state performance** of a stable system.


$$e(s)=\lim_{t\to\infty} e(t) = \lim_{s\to0} sE(s)$$

$$e(s)=\lim_{s\to0} {\frac{sR(s)}{1+G(s)}}$$



 **Unit Step Input**

the unit step input $R(s) = \frac{1}{s}$

steady state error of unit step input:
$e_{ss}(s)=\lim_{s\to0} {\frac{s}{1+G(s)}\frac{1}{s}}$

$e_{ss}(s)=\frac{1}{1+G(0)}$

Let's define **the position error constant** $K_p$

$K_p=\lim_{s\to0}{G(0)}=G(0)$

$\hspace{2cm}$$e_{ss}=\frac{1}{1+K_p}$

**For type 0 system:**

$K_p =\lim_{s\to0}{\frac{K(T_as+1)(T_bs+1)...(T_ms+1)}{(T_1s+1)(T_2s+1)...(T_ps+1)}}$

$\hspace{2cm}$$e_{ss}=\frac{1}{1+K}$

**For type 1 system:**

$K_p =\lim_{s\to0}{\frac{K(T_as+1)(T_bs+1)...(T_ms+1)}{s^N(T_1s+1)(T_2s+1)...(T_ps+1)}}=\infty\hspace{2cm} for N\geqslant1$

$\hspace{2cm}$$e_{ss}=0$

 **Unit Ramp Input**
 
 the unit ramp input $R(s) = \frac{1}{s^2}$
 
 steady state error of unit ramp input:
$e_{ss}(s)=\lim_{s\to0} {\frac{s}{1+G(s)}\frac{1}{s^2}}$

$e_{ss}(s)=\frac{1}{sG(s)}$

Let's define **the velocity error constant** $K_v$

$K_v=\lim_{s\to0}{sG(s)}$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_v}$

**For type 0 system:**

$K_v =\lim_{s\to0}{\frac{sK(T_as+1)(T_bs+1)...(T_ms+1)}{(T_1s+1)(T_2s+1)...(T_ps+1)}}=0$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_v}=\infty$

**For type 1 system:**

$K_v =\lim_{s\to0}{\frac{sK(T_as+1)(T_bs+1)...(T_ms+1)}{s(T_1s+1)(T_2s+1)...(T_ps+1)}}=K$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_v}=\frac{1}{K}$

**For type 2 system:**

$K_v =\lim_{s\to0}{\frac{sK(T_as+1)(T_bs+1)...(T_ms+1)}{s^N(T_1s+1)(T_2s+1)...(T_ps+1)}}=\infty\hspace{2cm} for N\geqslant2$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_v}=0$

**Unit Parabolic Input**
 
 the unit parabolic input $R(s) = \frac{1}{s^3}$
 
 steady state error of unit ramp input:
$e_{ss}(s)=\lim_{s\to0} {\frac{s}{1+G(s)}\frac{1}{s^3}}$

$e_{ss}(s)=\frac{1}{sG(s)}$

Let's define **the acceleration error constant** $K_a$

$K_v=\lim_{s\to0}{s^2G(s)}$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_a}$

**For type 0 system:**

$K_a =\lim_{s\to0}{\frac{s^2K(T_as+1)(T_bs+1)...(T_ms+1)}{(T_1s+1)(T_2s+1)...(T_ps+1)}}=0$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_a}=\infty$

**For type 1 system:**

$K_a =\lim_{s\to0}{\frac{s^2K(T_as+1)(T_bs+1)...(T_ms+1)}{s(T_1s+1)(T_2s+1)...(T_ps+1)}}=0$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_a}=\infty$

**For type 2 system:**

$K_a =\lim_{s\to0}{\frac{s^2K(T_as+1)(T_bs+1)...(T_ms+1)}{s^2(T_1s+1)(T_2s+1)...(T_ps+1)}}=K$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_a}=\frac{1}{K}$

**For type 3 system:**

$K_a =\lim_{s\to0}{\frac{s^2K(T_as+1)(T_bs+1)...(T_ms+1)}{s^N(T_1s+1)(T_2s+1)...(T_ps+1)}}=\infty\hspace{2cm} for N\geqslant3$

$\hspace{2cm}$$e_{ss}=\frac{1}{K_a}=0$

**Example** -Type 0 System

---
<figure>
<img src='https://live.staticflickr.com/65535/50080802351_594fbeda0e_b.jpg' />
<figcaption> $\hspace{6cm}$ Figure 6</figcaption>

A unity-feedback type 0 system is given as *Figure 6*.  

Let's find the unit step, ramp and parabolic input and observe the steady-state error.


```
time = np.linspace(0,50,5001)
input = np.ones(5001)
num = [1,1]
den = [1,3,4]
T,yout,xout = signal.lsim((num,den), input ,time)
#lsim command simulates output of a continuous-time linear system. 
#(num,den) = System's transfer function
#T = Time values for the output , yout = System response , xout = Time evolution of the state vector.
plt.plot(time,yout,label ="Step Response")
plt.plot(time,input, label ="Step Input")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xlabel("Time [s]")
plt.ylabel("Step Response")
plt.title("Step Response vs. Time")
plt.grid(True)
plt.show()

input = time
num = [1,1]
den = [1,3,4]
T,yout,xout = signal.lsim((num,den), input ,time)
plt.plot(time,yout)
plt.plot(time,input)
plt.plot(time,yout,label ="Ramp Response")
plt.plot(time,input, label ="Ramp Input")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xlabel("Time [s]")
plt.ylabel("Ramp Response")
plt.title("Ramp Response vs. Time")
plt.grid(True)
plt.show()

input = 1/2 * time**2
num = [1,1]
den = [1,3,4]
T,yout,xout = signal.lsim((num,den), input ,time)
plt.plot(time,yout)
plt.plot(time,input)
plt.plot(time,yout,label ="Parabolic Response")
plt.plot(time,input, label ="Parabolic Input")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xlabel("Time [s]")
plt.ylabel("Parabolic Response")
plt.title("Parabolic Response vs. Time")
plt.grid(True)
plt.show()
```
